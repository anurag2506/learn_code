Write a script that:

Loads an image.

Rotates it 45Â° about its center.

Expands the canvas so the entire image fits (no clipping).

Saves it as rotated.png.

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
p = '../image.png'
img = cv2.imread(p)
```

```python
def rotate_(img:np.array, angle:float):
    # cos_ = np.cos(np.deg2rad(angle))
    # sin_ = np.sin(np.deg2rad(angle))
    # rot_mat = np.array([[cos_, -sin_,0],[sin_, cos_,0],[0,0,0]])
    # rot_mat = rot_mat[:2,:]
    cx,cy = img.shape[0]//2, img.shape[1]//2
    rot_mat = cv2.getRotationMatrix2D((cx,cy),angle,scale=1.0)
    rotated = cv2.warpAffine(img, rot_mat,(img.shape[1],img.shape[0]))
    plt.imshow(rotated)
    
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def rotate_expand(img: np.array, angle: float):
    h, w = img.shape[:2]
    cx, cy = w // 2, h // 2
    rot_mat = cv2.getRotationMatrix2D((cx, cy), angle, 1.0)

    radians = np.deg2rad(angle)
    new_w = int(abs(h * np.sin(radians)) + abs(w * np.cos(radians)))
    new_h = int(abs(h * np.cos(radians)) + abs(w * np.sin(radians)))

    rot_mat[0, 2] += (new_w // 2) - cx
    rot_mat[1, 2] += (new_h // 2) - cy

    rotated = cv2.warpAffine(img, rot_mat, (new_w, new_h))
    
    plt.imshow(rotated)
    plt.axis("off")
    plt.show()

```

```python
rotate_expand(img,45)
```
