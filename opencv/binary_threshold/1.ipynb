```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
img = '../image.png'
image = cv.imread(img)
```

Thresholding:
If the pixel val is smaller than or rqual to the threshold, it is set to 0, otherwise set to a max value

Types:
1. Thresh_binary
2. Thresh_binary_inv
3. Thresh_trunc
4. Thresh_tozero
5. Thresh_tozero_inv
6. Thresh_mask
7. Thresh_otsu
8. Thresh_triangle
9.Thresh Dryrun

```python
def threshold(img: np.array):
    gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    ret,thresh1 = cv.threshold(img,128,255,cv.THRESH_BINARY)
    plt.imshow(thresh1)
    plt.show()
```

```python
threshold(image)
```
